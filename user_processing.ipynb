```python
import pandas as pd
import numpy as np
import json
import os
```

```python
os.chdir('users')
```

```python
pages = []
for file in os.listdir():
    print(file)
    page = pd.read_json(file)
    pages.append(page)
data = pd.concat(pages, ignore_index=True)
unique_data = data.drop_duplicates(subset=['steam_id'], keep='first', ignore_index=True)
```

```python
len(set(data['steam_id'])) == len(unique_data)
```

```python
page_size = len(unique_data)
users = []
for i in range(0, page_size):
    steam_id = unique_data['steam_id'][i]
    user_games = pd.DataFrame(unique_data['response'][i])
    size = len(user_games['games'])
    games = []
    for j in range(0, size):
        game = pd.DataFrame([user_games['games'][j]])
        games.append(game)
    user_games_df = pd.concat(games, ignore_index=True)
    user_games_df['steam_id'] = steam_id
    users.append(user_games_df)
users_df = pd.concat(users, ignore_index=True)
```

```python
users_df = users_df[['steam_id', 'appid', 'name', 'playtime_forever', 'playtime_2weeks', 'has_leaderboards', 'has_community_visible_stats', 'content_descriptorids']]
```
